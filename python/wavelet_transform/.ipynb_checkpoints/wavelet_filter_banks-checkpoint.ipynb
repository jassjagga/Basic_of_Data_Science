### Wavelet_transform

#### Filter Bank of a Discrete Wavelet 

```python
import pywt
import matplotlib.pyplot as plt
```

```python
wavelet =pywt.Wavelet("db7")
wavelet.filter_bank
```

```python
wavelet.dec_hi
```

```python
wavelet.dec_lo
```

```python
wavelet.dec_len
```

```python
[dec_lo,dec_hi,rec_lo,rec_hi]=wavelet.filter_bank


fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(2,2,figsize=(15,7))

ax1.stem(dec_lo)
ax1.set_title("Decomposition Low-Pass Filter")


ax2.stem(dec_hi)
ax2.set_title("Decomposition High-Pass Filter")

ax3.stem(rec_lo)
ax3.set_title("Reconstruction Low-Pass Filter")

ax4.stem(rec_lo)
ax4.set_title("Reconstruction High-Pass Filter")
```
