```python
import pandas as pd 
df1 =pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temp":[21,14,35],
})
df1
```

```python
df2 =pd.DataFrame({
    "city":["chicago","new york","orlando"],
    "humidity":[65,68,75],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")
df3
```

```python
df4 =pd.DataFrame({
    "city":["new york","chicago","orlando","baltimore"],
    "temp":[21,14,35,32],
})
df4
```

```python
df5 =pd.DataFrame({
    "city":["chicago","new york","sanfranciso"],
    "humidity":[65,68,71],
})
df5
```

```python
df6=pd.merge(df4,df5,on="city")
df6
```

```python
df6=pd.merge(df4,df5,on="city",how="outer")  
df6
```

```python
df6=pd.merge(df4,df5,on="city",how="left")  
df6
```

```python
df6=pd.merge(df4,df5,on="city",how="right")  
df6
```

```python
df6=pd.merge(df4,df5,on="city",how="outer", indicator="True")  
df6
```

```python
df9=pd.merge(df4,df5,on="city",how="left")  
df9
```

```python
df9=df9.fillna(71)
df9
```

```python
df5 =pd.DataFrame({
    "city":["chicago","new york","sanfranciso"],
    "temp":[21,14,35],
    "humidity":[65,68,71],
})
df5
```

```python
df10=pd.merge(df9,df5,on="city")
df10
```

```python
df10=pd.merge(df9,df5,on="city",suffixes=("_left","_right"))
df10
```
