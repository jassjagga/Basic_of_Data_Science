```python
import pandas as pd
import datetime
```

```python
df=pd.read_csv("TSLA.csv",parse_dates=["Date"])
df.head()
```

```python
df= df.drop("Date",axis=1)
df.head()
```

```python
holidays_list=[datetime.datetime(2021, 7, 1),datetime.datetime(2021, 9, 6),datetime.datetime(2021, 10, 11),
               datetime.datetime(2021, 12, 27),datetime.datetime(2021, 12, 28),datetime.datetime(2022, 1, 3),
              datetime.datetime(2022, 2, 21),datetime.datetime(2022, 4, 15),datetime.datetime(2022, 5, 23)]
holidays_list
```

```python
rng =pd.bdate_range(start="24-06-2021",end="24-06-2022",freq="C",holidays=holidays_list) #B means business days , C means custom string
rng
```

```python
df.set_index(rng,inplace =True)
df
```

```python
df.Close.plot()
```
