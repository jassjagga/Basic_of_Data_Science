# Missing Data 

### Topics 

1. Fillna to fill missing values wusing different ways
2. Interpolate to make guess on missing values using interpolation
3. Dropna to drop rows with missing values

```python
import pandas as pd
df = pd.read_csv("weather_data.csv", parse_dates=["day"])
#  type(df.day[0])  #to find out the type after parse_dates from string to timestamp
df.set_index('day',inplace=True)
df
```

### 1. fillna 

```python
new_df=df.fillna(0)
new_df
```

#### Fill spicific values to differenct rows or cloums

```python
new_df = df.fillna({
    'temp':0,
    'windspeed':0,
    'event':'No Event'
})
new_df
```

### carry forward the temp instead of 0 for good results (Forward Fill Method)

```python
new_df=df.fillna(method="ffill")
new_df
```

```python
new_df1=df.fillna(method="bfill") #backward filling the missing data
new_df1
```

```python
# axis in 
new_df2=df.fillna(method="bfill", axis="columns")  #filling the data columns wise 
new_df2
```

```python
new_df3=df.fillna(method="ffill",  limit=1) #only on numbr of value is forward filled like 32 next is still NaN
new_df3
```

### Interplot Method to get the intermediate value

```python
new_df4=df.interpolate()
new_df4
```

```python
# time accoridng in interpolate 
new_df5=df.interpolate(method="time")
new_df5
```

### 3. Drop Na Method

```python
new_df6= df.dropna() #Drop all the columns with NaN value
new_df6
```

```python
#if some rows have  data still missing but some rows have data so
new_df7= df.dropna(how="all") 
new_df7
```

```python
#if it at least has one value like NaN
new_df7= df.dropna(thresh=1) 
new_df7
```

```python
#if it at least has one value like NaN
new_df7= df.dropna(thresh=2) #at least 2 values to be need in this case
new_df7
```
