```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,53,48,47,49,46]
mint_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

## Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[19,2,40,3]
```

```python
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="Profit")
plt.legend()
```

```python
#here the bar's are one top of each other because the cpmany string has same location so thats by we have to convert in numbe string
#example like this 
xpos= np.arange(len(company))  #to genrate the string with company length
plt.xticks(xpos,company) #then puting the cpomany name instead of a string number place

plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")               
plt.bar(xpos-0.2,revenue,label="Revenue",width=0.4)
plt.bar(xpos+0.2,profit,label="Profit",width=0.4)
plt.legend()
```

```python
#here the bar's are one top of each other because the cpmany string has same location so thats by we have to convert in numbe string
#example like this 
xpos= np.arange(len(company))  #to genrate the string with company length
plt.yticks(xpos,company) #then puting the cpomany name instead of a string number place

plt.xlabel("revenue(bln)")
plt.title("US Tech Stocks")               
plt.barh(xpos-0.2,revenue,label="Revenue",height=0.4)
plt.barh(xpos+0.2,profit,label="Profit",height=0.4)
plt.legend()
```

### Histograms

```python
blood_suger=[113,85,90,150,149,88,93,115,132,80,77,82,129]
plt.hist(blood_suger)
```

```python
# bin or buket is like range of 10 in this we need range for three things low high or normal blood pressure 
plt.hist(blood_suger, bins=3, rwidth=0.95) # in this a self define range 
```

```python

plt.hist(blood_suger, bins=[80,100,125,150], rwidth=0.95,color='g') #we define the range in these bins 
```

```python
blood_suger_men=[113,85,90,150,149,88,93,115,132,80,77,82,129]
blood_suger_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.ylabel('Total no of Patients')
plt.xlabel("Suger Range")
plt.title("Blood Suger Analysis")
plt.hist([blood_suger_men,blood_suger_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=["Mens","Women"])
plt.legend()
plt.grid()
```

```python
plt.xlabel('Total no of Patients')
plt.ylabel("Suger Range")
plt.title("Blood Suger Analysis")
plt.hist([blood_suger_men,blood_suger_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],
         label=["Mens","Women"],orientation='horizontal')
plt.legend()
```

### Pie Chart

```python
exp_vals=[1400,600,300,410,250]
exp_labels=["Home Rent","Food", "Phone/Internet Bill","car","other utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal") # in case to make it tilt or strage circle
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.1f%%', shadow=True)
plt.show()
```

```python
plt.axis("equal") # in case to make it tilt or strage circle
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.1f%%', shadow=True,explode=[0,0.1,0.2,0,0])
plt.show()
```

```python
plt.axis("equal") # in case to make it tilt or strage circle
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.1f%%', shadow=True,explode=[0,0.1,0.2,0,0],startangle=180) 
                                                                                                #To change the aggle
plt.show()
```
