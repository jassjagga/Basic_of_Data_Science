### K Means Cluster Algorithm 

1. Supervise Learning : - In supervise learning we will super wise the learning process of the model. For example in vrious methos like LogisticRegression or Randon Forest etc methonds we have the Target values with the data to so model can learn from that data. and then we test the model by sample testing data set with a target values.
2. Unsupervise Learning: - In unsupervise learning the target or output data is note given with the data. Only the simple data is provided and we have to find the meaning of that data without any target or output data. means model has to learn from the data and find the meaning of data without any expected output.
